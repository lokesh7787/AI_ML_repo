```python
import pandas as pd
from datetime import datetime
from sqlalchemy import create_engine
```

```python
#collecting dataset
df_customer = pd.read_csv("Customers.csv",encoding='latin-1')
df_ExchangeRates = pd.read_csv("Exchange_Rates.csv",encoding='latin-1')
df_Products = pd.read_csv("Products.csv",encoding='latin-1')
df_Sales = pd.read_csv("Sales.csv",encoding='latin-1')
df_Store = pd.read_csv("Stores.csv",encoding='latin-1')
```

```python
#checking null value for customer data set
df_customer.info()
```

```python
# drop null value
df_customer.dropna(axis=0,inplace=True)
```

```python
#checking customer dataset after removing null
df_customer.info()
```

```python
# checking null value for Exchange dataset
df_ExchangeRates.info()
```

```python
#checking null value Products
df_Products.info()
```

```python
# checking null value Sales dataset
df_Sales.info()
```

```python
#checking null value in Store dataset
df_Store.info()
```

```python
#clearing null value in Store dataset
df_Store.dropna(axis=0, inplace=True)
```

```python
# after clearing null checking Store dataset 
df_Store.info()
```

```python
#Feature enginering 

#currency into number in Products dataset
df_Products['Unit Cost USD'] = df_Products['Unit Cost USD'].replace("[$,]", "", regex=True).astype(float)
df_Products['Unit Price USD'] = df_Products['Unit Price USD'].replace("[$,]", "", regex=True).astype(float)
```

```python
#checking Products dataset
df_Products.head()
```

```python
# customers  creating new column Age 

def age(dob):
  dob = datetime.strptime(dob,"%m/%d/%Y").date()
  today = datetime.today()
  return today.year - dob.year - ((today.month,today.day) < (dob.month,dob.day))  

df_customer['Age'] = df_customer['Birthday'].apply(age)

# formating to standardize date formate Customer , Sales and Exchange datasets
df_customer['Birthday'] = pd.to_datetime(df_customer['Birthday'],errors = 'coerce')
df_Sales['Order Date'] = pd.to_datetime(df_Sales['Order Date'],errors = 'coerce')
df_Sales['Delivery Date'] = pd.to_datetime(df_Sales['Delivery Date'],errors = 'coerce')
df_ExchangeRates['Date'] = pd.to_datetime(df_ExchangeRates['Date'],errors="coerce")
```

```python
#checking after Adding column Age and Birday column standardize date format in customer dataser
df_customer.head()
```

```python
#checking after Delivery date and Order Date columns in Sales dataset
df_Sales.head()
```

```python
# checking after date in ExchangeRates dataset
df_ExchangeRates.head()
```

```python
# here calculating Profit Margin and adding new column in Products dataset
df_Products['Profit Margin (%)'] = ((df_Products['Unit Price USD'] - df_Products['Unit Cost USD']) / df_Products['Unit Price USD'] ) * 100
```

```python
# checking after calculation Products dataset
df_Products.head()
```

```python
# adding new column delivery status (Deliverd / Not Delivered) in Sales dataset

df_Sales['Delivery Status'] = df_Sales['Delivery Date'].apply(lambda x :'Not Delivered' if pd.isna(x) else 'Delivered')

```

```python
# checking after creating new flag column

df_Sales.head()
```

```python
# merging sales data with customer data

df_mergedData = df_Sales.merge(df_customer,how='inner',on='CustomerKey')
```

```python
df_mergedData.head()
```

```python
# merging with products dataset
df_mergedData = df_mergedData.merge(df_Products,how='inner',on='ProductKey')
```

```python
# checking after merging with products dataset 
df_mergedData.head()
```

```python
# merging with Stores dataset 

df_mergedData = df_mergedData.merge(df_Store,how='inner',on='StoreKey')
```

```python
# checking after merging with Store dataset
df_mergedData.head()
```

```python
# merging with Exchange rate dataset

df_mergedData = df_mergedData.merge(df_ExchangeRates,how='inner',left_on=['Order Date','Currency Code'] ,right_on=['Date','Currency'])
```

```python
#checking after merging with exchanges rates dataset
df_mergedData.head()
```

```python
# removing duplicate coumn Date , currency state country  
df_mergedData = df_mergedData.drop(['Date','Currency','State_y','Country_y'], axis=1)

#rename cloumn State and country
df_mergedData.rename(columns={'State_x':'State','Country_x':'Country'}, inplace=True)
```

```python
#checking after removing 2 columns
df_mergedData.head()
```

```python
# calculating sales amount 
df_mergedData['Sales Amount'] = ((df_mergedData['Quantity'] * df_mergedData['Unit Price USD']) * df_mergedData['Exchange'])
```

```python
#checking sales amount calcualtion
df_mergedData[['Quantity','Unit Price USD','Exchange','Sales Amount']].head()
```
